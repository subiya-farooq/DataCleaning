```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

country = {
        'country': ['India', 'Japan', 'Nepal', 'Bihar'],
        'Population': [8.5, 4.0, 2.7, 5.5],
        'GDP': [98, 34, 22, 100],
        'Area':[50,10,20,100]
}

df= pd.DataFrame(country)
df
```

```python
plt.plot(df['country'], df['Population'])
```

```python
x= df['country']
y= df['Population']

plt.plot(x,y, marker='*')

plt.xlabel('country')
plt.ylabel('Population')
```

```python
x= df['country']
y= df['Population']
q=df['GDP']
r=df['Area']
plt.scatter(x,y, marker='*')

plt.scatter(q,r)
plt.plot(x,q)
```

```python
plt.pie(q,labels=x)
```
