```python
import numpy as np
import pandas as pd

df = pd.read_csv('dirty_cafe_sales.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Item'].unique()
```

```python
df['Item'] = df['Item'].fillna(df['Item'].mode()[0])
```

```python

df['Quantity'] = df['Quantity'].str.replace(",", "", regex=False)
df['Quantity'] = pd.to_numeric(df['Quantity'], errors="coerce")
```

```python
df['Quantity'].unique()
```

```python
df['Quantity'].skew()
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
```

```python
df['Price Per Unit'] = df['Price Per Unit'].str.replace(",", "", regex=False)
df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors="coerce")
```

```python
df['Price Per Unit'].unique()
```

```python
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].mean())
```

```python
df.info()
```

```python
df = df.drop( columns=['Total Spent'])
df
```

```python
df['Total Spent'] = df['Quantity'] * df['Price Per Unit']
df
```

```python
df['Payment Method'].unique()
```

```python
df['Payment Method'] = df['Payment Method'].replace('UNKNOWN', df['Payment Method'].mode()[0])
df
```

```python
df['Payment Method'] = df['Payment Method'].fillna(df['Payment Method'].mode()[0])
df['Location'] = df['Location'  ].fillna(df['Location'].mode()[0])
df
```

```python
df['Payment Method'].unique()
```

```python

df = pd.get_dummies(df, columns=['Payment Method'], drop_first=True)
df
```
